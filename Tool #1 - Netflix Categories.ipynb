```python
word_count_dict = {}
word_list = []
ignore_list = ["for", "2", "12", "11", "to", "10", "on", "5", "0", "4", "8", "&", "7", "and"]
all_words = [tuple(word_list.append(c) for c in cat.lower().split()) for cat in netflix_cats["categories"]]
word_set = set(word_list)


for keyword in word_set:
    if keyword in ignore_list:
        pass
    else:
        word_count = 0
        for cat in word_list:
            if keyword in cat:
                word_count += 1
            else:
                pass
        word_count_dict[keyword] = word_count

for x in sorted([(key,val) for key,val in word_count_dict.items()], key=lambda x: x[1], reverse= True):
    print(x)
```
